```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Stripplot

```python
_ = sns.catplot(tips, x='day', y='total_bill')
```

Swarmplot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

```python
# W odróżnieniu do stripplotu, tam, gdzie jest więcej wartości nagromadzonych, wykres się rozsuwa.
```

Boxplot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box')
```

```python
# Kreska w pudełku określa medianę, wąsy to pierwszy i trzeci kwartyl. Im wyższy prostokąt, tym bardziej rozrzucone wartości. Kropki oznaczają przypadki odstające.
```

Bar

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

```python
# Średnia dla danej katgorii. Przedział ufności - kreska na kolumnie.
```

Countplot

```python
_ = sns.catplot(tips, x='day', kind='count')
```
