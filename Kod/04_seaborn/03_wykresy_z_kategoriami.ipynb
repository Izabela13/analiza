```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Stripplot

```python
_ = sns.catplot(tips, x='day', y='total_bill')
```

Swarmplot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

```python
# W odrnieniu do stripplotu, tam, gdzie jest wicej wartoci nagromadzonych, wykres si rozsuwa.
```

Boxplot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box')
```

```python
# Kreska w pudeku okrela median, wsy to pierwszy i trzeci kwartyl. Im wyszy prostokt, tym bardziej rozrzucone wartoci. Kropki oznaczaj przypadki odstajce.
```

Bar

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

```python
# rednia dla danej katgorii. Przedzia ufnoci - kreska na kolumnie.
```

Countplot

```python
_ = sns.catplot(tips, x='day', kind='count')
```
