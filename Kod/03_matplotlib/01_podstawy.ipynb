### Import biblioteki matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Pierwszy wykres

```python
_ = plt.plot([0, 1], [0, 1])  # dwa parametry funkcji liniowej - dla osi "x" (x1=0, x2=0) i osi "y" (y1=1, y2=1)
```

```python
x = np.arange(-1, 1.1, 0.1)
y = x ** 2
```

```python
_ = plt.plot(x, y)
```

Tytu wykresu

```python
plt.plot(x, y)
_ = plt.title('Wykres x * x')
```

Dwie funkcje na jednym wykresie

```python
x1 = np.arange(-1, 1.1, 0.1)  # punkty na osi "x"
y1 = x1 ** 3  # punkty na osi "y"
```

```python
# Tworzenie wykresu I funkcji
plt.plot(x, y)

# Tworzenie wykresu II funkcji
_ = plt.plot(x1, y1)
```

```python
# I funkcja - kwadratowa (kolor niebieski)
# II funkcja - do potgi 3 (kolor pomaraczowy)
```

Etykiety

```python
plt.plot(x, y, label='x ^ 2')
plt.plot(x1, y1, label='x ^ 3')
_ = plt.legend()
```

```python
# Tworzymy jeden obszar wykresu, ktry bdzie zawiera dwa wykresy - po jednej i po drugiej stronie
# Korzystamy z funkcji "subplot"
```

Wiele wykresw

```python
# Definiujemy, jaki sublot chcemy utworzy - nasz obszar to jeden wiersz i dwie kolumny i pole, w ktrym chcemy tworzy wykres
plt.subplot(1, 2, 1) # element nr 1
plt.plot(x, y)

plt.subplot(1, 2, 2) # element nr 2
_ = plt.plot(x1, y1)
```

Wykres supkowy

```python
x2 = np.arange(1, 6) # kolejne kolumny (5)
y2 = [10, 20, 40, 15, 99]
```

```python
_ = plt.bar(x2, y2)
```
